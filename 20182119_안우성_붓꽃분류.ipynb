```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
```

## 1) 붓꽃 데이터를 70% 학습, 30% 테스트 분리

```python
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

## 2) 데이터 표준화(standardization)

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 3) 각 모델의 정확률, 혼동 행렬

### 3-1) K - NN (K = 5)

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
knn_cm = confusion_matrix(y_test, knn_pred)
```

```python
print("K-NN 모델의 정확도:", knn_acc)
print("K-NN 모델의 혼동행렬:\n", knn_cm)
```

###  3-2) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
logreg_pred = logistic_regression.predict(X_test)
logreg_acc = accuracy_score(y_test, logreg_pred)
logreg_cm = confusion_matrix(y_test, logreg_pred)
```

```python
print("Logistic Regression 모델의 정확도:", logreg_acc)
print("Logistic Regression 모델의 혼동행렬:\n", logreg_cm)
```

### 3-3) Decision-Tree : max_depth = 4, criterion = gini

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree_gini = DecisionTreeClassifier(max_depth=4, criterion='gini')
decision_tree_gini.fit(X_train, y_train)
dt_gini_pred = decision_tree_gini.predict(X_test)
dt_gini_acc = accuracy_score(y_test, dt_gini_pred)
dt_gini_cm = confusion_matrix(y_test, dt_gini_pred)
```

```python
print("결정 트리 max_depth=4,criterion=gini 모델의 정확도:", dt_gini_acc)
print("결정 트리 max_depth=4,criterion=gini 모델의 혼동행렬:\n", dt_gini_cm)
```

### 3-4) Decision-Tree : max_depth = 4, criterion = entropy

```python
dt_entropy = DecisionTreeClassifier(max_depth=4, criterion='entropy')
dt_entropy.fit(X_train, y_train)
dt_entropy_pred = dt_entropy.predict(X_test)
dt_entropy_acc = accuracy_score(y_test, dt_entropy_pred)
dt_entropy_cm = confusion_matrix(y_test, dt_entropy_pred)
```

```python
print("결정 트리 max_depth=4,criterion=entropy 모델의 정확도:", dt_entropy_acc)
print("결정 트리 max_depth=4,criterion=entropy 모델의 혼동행렬:\n", dt_entropy_cm)
```

### 3-5)LinerSVM


```python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
svm_cm = confusion_matrix(y_test, svm_pred)
```

```python
print("Linear SVM 모델의 정확도:", svm_acc)
print("Linear SVM 모델의 혼동행렬:\n", svm_cm)
```

# 4) 앙상블의 정확률

```python
from sklearn.ensemble import VotingClassifier

ensemble = VotingClassifier(estimators=[('knn', knn), ('logreg', logistic_regression), ('dt_gini', decision_tree_gini), ('dt_entropy', dt_entropy), ('svm', svm)], voting='hard')
ensemble.fit(X_train, y_train)
ensemble_pred = ensemble.predict(X_test)
ensemble_acc = accuracy_score(y_test, ensemble_pred)
ensemble_cm = confusion_matrix(y_test, ensemble_pred)
```

```python
print("앙상블 모델의 정확도:", ensemble_acc)
print("앙상블 모델의 혼동행렬:\n", ensemble_cm)
```
